```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/hepatitis (1).csv")
data
```

```python
not_good= data['histology']<=1
is_good= data['histology']>=2
print(not_good.value_counts())
print(is_good.value_counts())
data['histology'][not_good]=0
data['histology'][is_good]=1
data['histology'].value_counts()
print(data)
```

```python
data['class'][not_good]=0
data['class'][is_good]=1
data['sex'][not_good]=0
data['sex'][is_good]=1
data['steroid'][not_good]=0
data['steroid'][is_good]=1
data['antivirals'][not_good]=0
data['antivirals'][is_good]=1
data['fatigue'][not_good]=0
data['fatigue'][is_good]=1
data['malaise'][not_good]=0
data['malaise'][is_good]=1
data['anorexia'][not_good]=0
data['anorexia'][is_good]=1
data['liver_big'][not_good]=0
data['liver_big'][is_good]=1
data['liver_firm'][not_good]=0
data['liver_firm'][is_good]=1
data['spleen_palable'][not_good]=0
data['spleen_palable'][is_good]=1
data['spiders'][not_good]=0
data['spiders'][is_good]=1
data['ascites'][not_good]=0
data['ascites'][is_good]=1
data['varices'][not_good]=0
data['varices'][is_good]=1
```

```python
X=data.drop('histology',axis=1)
y=data['histology']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
import tensorflow as tf
```

```python
model= tf.keras.Sequential([
                            tf.keras.layers.Dense(19, activation='relu'),
                            tf.keras.layers.Dense(10, activation='relu'),
                            tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss= tf.keras.losses.binary_crossentropy,
              optimizer= tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics= [tf.keras.metrics.BinaryAccuracy(name='accuracy'),
                        tf.keras.metrics.Precision(name='precision'),
                        tf.keras.metrics.Recall(name='a=recall')
              ]
              )
```

```python
history= model.fit(X_train, y_train, epochs= 100, verbose=0)
```

```python
model.evaluate(X_test,y_test)
```

```python
pd.DataFrame(history.history).plot()
```
